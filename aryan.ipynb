```
# SHALE VERY IMPORTANT IN DETERMINING PAY ZONE I.E SHALE OIL AND SHALE GAS, ESXPICIALLY IN CONVEMTIONAL RES
#SHALE VOLUME CANE BE DETERMINED USING GAMMA RAY LOGS
# LESS SHALE = BETTER RESERVOIR
# MORE SHALE = BETTER CAP ROCK

# shale volume = (GR-GRmin)/(GRmax - GRmin)

def fillna(log):
  rem = log[~np.isnan(log)]
  mean =np.mean(rem)
  log[np.isnan(log)] = mean
  return log

upd_gr= fillna(gr)
```

```
upd_gr
```

```
GRmax=np.max(upd_gr)
GRmin=np.min(upd_gr)

VSH = (upd_gr-GRmin)/(GRmax-GRmin)
```

```
VSH
```

```
plt.figure(figsize=(7,7))

plt.subplot(1,2,1)
plt.plot(gr,depth,color='black')
plt.ylim(max(depth),min(depth))
plt.title('GR')

plt.subplot(1,2,2)
plt.plot(VSH,depth,color='purple')
plt.ylim(max(depth),min(depth))  #specifying limits important
plt.title('VSH')
```
